```python
import re


def choice():
    print(
        "Please choose what would you like to do : \n Type 1 for Login \n Type 2 for Registration \n type 3 for "
        "forgot password")
    user_responce = int(input())
    if user_responce == 1:
        user_login()
    elif user_responce == 2:
        user_registration()

    elif user_responce == 3:
        forgot_password()


def forgot_password():
    f = open("User_Data.txt", 'r')
    info = f.read()
    info = info.split()
    username = input("Enter your username")
    password = input("Enter your new password")
    if username in info:
        index = info.index(username) + 1
        info[index] = password
        f = open("User_Data.txt", 'w')
        info = " ".join(str(item) for item in info)
        f.write(info)
        print("password changed")
    else:
        pass


def store_username_pass(username, password):
    f = open("User_Data.txt", 'w+')
    info = f.read()
    if username in info:
        print("Name Unavailable. Please Try Again")
    f.close()
    f = open("User_Data.txt", 'w')
    info = info + " " + username + " " + password
    f.write(info)


# username validation
def check_username(username):
    pattern = re.compile(r'(?=^.{6,30}$)(^[a-zA-Z][a-zA-Z0-9.,$;]*@[a-zA-Z]+\.(com|in))')
    matches = pattern.finditer(username)
    return matches


# password validation
def password_check(password):
    reg = "^(?=.*[a-z])(?=.*[A-Z])(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{5,16}$"
    pat = re.compile(reg)
    mat = re.search(pat, password)
    return mat


def user_login():
    f = open("User_Data.txt", 'r')
    info = f.read()
    info = info.split()
    username = input("Enter your username : ")
    password = input("Enter your password : ")
    if username in info:
        index = info.index(username) + 1
        usr_password = info[index]
        if usr_password == password:
            print("\n you are logged in\n")
            print("Welcome Back, " + username)
        else:
            print("Password entered is wrong")
    else:
        print("Name not found. Please Sign Up.")
        return user_registration()


# if username and password valid store it in text file
def user_registration():
    print("Please Provide")
    username = str(input(" username : "))
    flag1 = check_username(username)
    if flag1:
        print("Username is valid.")
    else:
        print("username invalid !!")
        print("Renter username", check_username())

    password = str(input("Password : "))
    flag2 = password_check(password)
    if flag2:
        print("Password is valid.")
        print("Sign up is successful")
        print("login to your account")
        store_username_pass(username, password)
        user_login()
    else:
        print("Password invalid !!")
        print("Renter password", user_registration())


if __name__ == "__main__":
    choice()
```
